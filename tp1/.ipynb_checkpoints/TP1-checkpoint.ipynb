```python
%load_ext autoreload

%autoreload 2

import tp1_functions
```

```python
N = 1000
K = 4000
```

For every value k between 1 and K, we create a hash table of size N and fill it with k elements. Then, for each of these hash tables, we calculate the number of key for which there is exactly 1 element. Finally, we plot this number with respect to k.

```python
empirical_values = tp1_functions.plot_number_of_no_collisions(N, K)
```

We can model this curve with the function :  $f(k, N) = k(1 - \frac{1}{N})^{k - 1}$

```python
theoretical_values = tp1_functions.plot_empirical_and_theoretical_no_collisions(N, K, empirical_values)
```

We can generalise this by plotting the number of entries in the hashtable with exactly p elements. The theoretical expression of this value is : $f(k,N,p) = \binom{k}{p}\frac{1}{N^{p-1}}(1 - \frac{1}{N})^{k - p}$

```python
empirical_values, theoretical_values = tp1_functions.plot_empirical_and_theoretical_p_collisions(N, K, 2)
```

Moreover, we can plot the number of entries in the hash table with strictly more than 1 element, ie the number of collisions in the hash table. This value can be theoretically expressed as : $f(k, N) = N - N(1 - \frac{1}{N})^{k} - k(1 - \frac{1}{N})^{k-1}$

```python
empirical_values, theoretical_values = tp1_functions.plot_empirical_at_least_one_collision(N, K)
```

Next, we fill a dictionary of unique words contained in a file

```python
dictionary = tp1_functions.get_dictionary_set('../data/20ng.utf8.txt')
print(len(dictionary))
```

Then, we define a 'bad' hash function for strings :

```python
def bad_hash_function( word, k ):
    val = 58458454215852
    for c in word:
        val *= (ord(c) + 1)
    return val % k
```

Finally, we can plot the number of collisions for every hash table of size n, given that we fill the hash table with values from the dictionary, hashed with the bad hash function

```python
tp1_functions.plot_hash_table_dictionary(dictionary, bad_hash_function, 1000000, 100000, 2000000)
```
